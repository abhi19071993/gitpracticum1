Considering 100000 rows of data and the columns productID, overall, reviewText and reviewerID for building our recommender system

```python
df_tmp = df[1:100000]
df_work = df_tmp[['productID','overall','reviewText','reviewerID']].copy()
```

Finding the number of unique categories and checking null/dummy values for each of the working data.

```python
for col_name in df_work.columns:
    if df_work[col_name].dtypes == 'object':
        unique_cat = len(df_work[col_name].unique())
        print("'{col_name}' has {unique_cat} unique categories".format(col_name=col_name, unique_cat=unique_cat))
```

```python
# Checking for dummy values in the data set

df_work.isnull().sum().sort_values(ascending=False).head()
```
